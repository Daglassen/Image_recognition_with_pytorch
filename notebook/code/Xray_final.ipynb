# Load data for Pytorch

```python
class ChestDataset(Dataset) :
    def __init__(self,path,train=True,transform=None,labelmap=None):
        self.train=train
        self.data_path = path['image_path']
        #self.categories = path['category']
        self.transform = transform
        self.labelmap = labelmap
    
    def __len__(self) :
        return len(self.data_path)
    
    def __getitem__(self,idx) :
        img_path = self.data_path[idx]
        image = Image.open(img_path).convert('RGB')
        
        
        if re.search('/train/(.*)/',img_path) :
            label = re.search('/train/XR_(.*)/patient',img_path).group(1)
            
            
        elif re.search('/valid/(.*)/',img_path) :
            label = re.search('/valid/XR_(.*)/patient',img_path).group(1)
            
        label = self.labelmap[label]
        label = torch.as_tensor(label)
        
        if self.transform :
            image = self.transform(image)
        
        #sample = {'image' :image ,'label':label}
        sample = list((image, label))
        
        return sample
    
```

```python
#create transforms object
#resize so all images have same size 
#totensor - convert image to tensor 

transform = transforms.Compose([transforms.Resize((224,224)),
                               transforms.ToTensor(),
                               transforms.Normalize([0.485, 0.456, 0.406],[0.229, 0.224, 0.225])])
```

# Define a Convolutional Neural Network

```python
#conv2d(in_channels, out_channels, kernel_size)
#MaxPool2d(kernal_size, stride)
#Linear(in_channels, out_channels)


class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(3, 16, 5) #output shape  
        self.pool = nn.MaxPool2d(2, 2) #output shape 
        self.conv2 = nn.Conv2d(16, 32, 5)         # (32 x 53 x 53)
        self.conv3 = nn.Conv2d(32, 64, 5)         # (128 x 25 x 25)
        self.conv4 = nn.Conv2d(64, 128, 5)        # (128 x 10 x 10)
        

        self.fc1 = nn.Linear(128 * 10 * 10, 2048)    #  (32 * 53 * 53, 520): two layers, (64 * 25 * 25, 2048): 3 layers, (128 * 10 * 10, 2048): 4 layers
        self.fc2 = nn.Linear(2048, 1024)
        self.fc3 = nn.Linear(1024, 512)
        self.fc4 = nn.Linear(512, 84)
        self.fc5 = nn.Linear(84, 8)
        
        # Define proportion or neurons to dropout
        self.dropout = nn.Dropout(0.2)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))
        x = self.pool(F.relu(self.conv4(x)))
        #print(x.shape)       #  used for debugging when adding more layers
        x = torch.flatten(x, 1) # flatten all dimensions except batch
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = F.relu(self.fc4(x))
        x = self.fc5(x)
        return x


net = Net().to(device) 
print(net)
#train(net, NUM_EPOCHS)
```

```python
#define loss function and optimizer 
criterion = nn.CrossEntropyLoss().to(device) 
optimizer = optim.Adam(net.parameters(), lr=0.001)
```

# Train Model

```python
def my_plot(epochs, loss):
    plt.plot(epochs, loss)
```

```python
NUM_EPOCHS = 12
#NUM_EPOCHS = 4 

def train(model, NUM_EPOCHS):
    print("Using device:", device)
    loss_vals=  []

    for epoch in range(NUM_EPOCHS):  # loop over the dataset multiple times
    
        running_loss = 0.0
        epoch_loss= []        
        #running_corrects = 0.0

        for i, data in enumerate(trainloader, 0):
            # get the inputs; data is a list of [inputs, labels]

            #inputs = data['image'].to(device)  
            #labels = data['label'].to(device)
            
            inputs = data[0].to(device) 
            labels = data[1].to(device)

            # zero the parameter gradients
            optimizer.zero_grad()

            # forward + backward + optimize
            outputs = net(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            epoch_loss.append(loss.item())
            optimizer.step()
        
            # print statistics
            running_loss += loss.item()
            if i % 2000 == 1999:    # print every 2000 mini-batches 
                print(f'[{epoch + 1}, {i + 1:5d}] loss: {running_loss / 2000:.3f}')
                running_loss = 0.0
        
        
        loss_vals.append(sum(epoch_loss)/len(epoch_loss))
        
    # plotting
    print('Finished Training')
    my_plot(np.linspace(1, NUM_EPOCHS, NUM_EPOCHS).astype(int), loss_vals)
```

```python
train(net, NUM_EPOCHS)
```

# Translate CNN to BNN using Laplace Approximation

```python
'''
#load pre trained model 

#PATH = './xray_cnn_8_epoch.pth'
PATH = 'xray_cnn_12_epoch.pth'
net = Net().to(device)
net.load_state_dict(torch.load(PATH))
'''
```

Our CNN is trained via maximum a-posteriori(MAP) inference. We find the MAP estimate of the weights. A maximum a-posteriori estimation is an estimation of an unknown quantity that equals the mode of the posterior distibution. 

$\theta_{\text{MAP}} = \text{arg max}_{\theta} p(\theta \vert \mathcal{D}) = \text{arg max}_{\theta} \text{log } p(\mathcal{D} \vert \theta) + \text{log } p(\theta)$ 

The MAP estimate $\theta_{\text{MAP}}$ represents a single point estimate in the parameter space and hence it ignores the uncertainty inherent in the parameters of the network. Translating our CNN to a Bayesian Neural Network will attempt to capture this uncertainty by inferring the full posterior disbritution $p(\theta \vert \mathcal{D})$. 

Let $\mathcal{D}$ represent the classification data set. Laplace approximations use a second order expansion of $\mathcal{L}$ around $\theta_{MAP}$ to construct a gaussian approximation to the poseriror distribution $p(\theta \vert \mathcal{D})$ : 

$p(\theta \vert \mathcal{D}) \approx \mathcal{N} (\theta; \theta_{\text{MAP}}, \Sigma)$

```python
#prediction type is classiciation 
#want distribution of last layer of weights 
#create multiple laplace flavors 

la_diag = Laplace(net, 'classification', subset_of_weights = 'last_layer', hessian_structure = 'diag')
la_full = Laplace(net, 'classification', subset_of_weights = 'last_layer', hessian_structure = 'full')
la_kron = Laplace(net, 'classification', subset_of_weights = 'last_layer', hessian_structure = 'kron')
```

```python
"""
Created on Thu Oct 26 11:06:51 2017

@author: Utku Ozbulak - github.com/utkuozbulak
"""

model = net

class CamExtractor():
    """
        Extracts cam features from the model
    """
    def __init__(self, model, target_layer):
        self.model = model
        self.target_layer = target_layer
        self.gradients = None

    def save_gradient(self, grad):
        self.gradients = grad

    def forward_pass_on_convolutions(self, x):
        """
            Does a forward pass on convolutions, hooks the function at given layer
        """
        conv_output = None
        for module_pos, module in self.model.features._modules.items():
            x = module(x)  # Forward
            if int(module_pos) == self.target_layer:
                x.register_hook(self.save_gradient)
                conv_output = x  # Save the convolution output on that layer
        return conv_output, x

    def forward_pass(self, x):
        """
            Does a full forward pass on the model
        """
        # Forward pass on the convolutions
        conv_output, x = self.forward_pass_on_convolutions(x)
        x = x.view(x.size(0), -1)  # Flatten
        # Forward pass on the classifier
        x = self.model.classifier(x)
        return conv_output, x
        


class GradCam():
    """
        Produces class activation map
    """
    def __init__(self, model, target_layer):
        self.model = model
        self.model.eval()
        # Define extractor
        self.extractor = CamExtractor(self.model, target_layer)

    def generate_cam(self, input_image, target_class=None):
        # Full forward pass
        # conv_output is the output of convolutions at specified layer
        # model_output is the final output of the model (1, 1000)
        conv_output, model_output = self.extractor.forward_pass(input_image)
        if target_class is None:
            target_class = np.argmax(model_output.data.numpy())
        # Target for backprop
        one_hot_output = torch.FloatTensor(1, model_output.size()[-1]).zero_()
        one_hot_output[0][target_class] = 1
        # Zero grads
        self.model.features.zero_grad()
        self.model.classifier.zero_grad()
        # Backward pass with specified target
        model_output.backward(gradient=one_hot_output, retain_graph=True)
        # Get hooked gradients
        guided_gradients = self.extractor.gradients.data.numpy()[0]
        # Get convolution outputs
        target = conv_output.data.numpy()[0]
        # Get weights from gradients
        weights = np.mean(guided_gradients, axis=(1, 2))  # Take averages for each gradient
        # Create empty numpy array for cam
        cam = np.ones(target.shape[1:], dtype=np.float32)
        # Multiply each weight with its conv output and then, sum
        for i, w in enumerate(weights):
            cam += w * target[i, :, :]
        cam = np.maximum(cam, 0)
        cam = (cam - np.min(cam)) / (np.max(cam) - np.min(cam))  # Normalize between 0-1
        cam = np.uint8(cam * 255)  # Scale between 0-255 to visualize
        cam = np.uint8(Image.fromarray(cam).resize((input_image.shape[2],
                       input_image.shape[3]), Image.Resampling.LANCZOS))/255
        return cam


if __name__ == '__main__':
    # Get params
    target_example = 0  # 0: hand, 1:wrist, 2:shoulder 
    (original_image, prep_img, target_class, file_name_to_export, pretrained_model) =\
        get_example_params(target_example)
    # Grad cam
    grad_cam = GradCam(pretrained_model, target_layer=4)
    # Generate cam mask
    cam = grad_cam.generate_cam(prep_img, target_class)
    # Save mask
    save_class_activation_images(original_image, cam, file_name_to_export)
    print('Grad cam completed')
```
